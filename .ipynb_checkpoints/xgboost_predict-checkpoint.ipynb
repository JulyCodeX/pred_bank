```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score, KFold, GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from collections import Counter
from matplotlib import pyplot as plt
import xgboost as xgb
from sklearn.metrics import roc_auc_score
%matplotlib inline
```

```python
dataset_train = pd.read_csv('train_set.csv', delimiter=',')
dataset_train.head(5)
```

```python
dataset_test = pd.read_csv('test_set.csv', delimiter=',')
dataset_test.head(5)
```

```python
dataset_1 = dataset_train
dataset_2 = dataset_test
```

```python
dataset_1.dtypes
```

```python
for col in dataset_1.columns[dataset_1.dtypes == 'object']:
    lh = LabelEncoder()
    lh.fit(dataset_1[col])
    dataset_1[col] = lh.transform(dataset_1[col])
dataset_1.head()
```

```python
dataset_1.describe()
```

```python
for col in dataset_2.columns[dataset_2.dtypes == 'object']:
    lh = LabelEncoder()
    lh.fit(dataset_2[col])
    dataset_2[col] = lh.transform(dataset_2[col])
dataset_2.head()
```

```python
dataset_all = pd.concat([dataset_1, dataset_2])
std_scaler = StandardScaler()
dataset_all[['age','job','balance', 'duration', 'campaign', 'pdays', 'previous','day','month']] = std_scaler.fit_transform(dataset_all[['age','job','balance', 'duration', 'campaign', 'pdays', 'previous','day','month']])
dataset_all.head()
```

```python
dataset_train_columns = list(set(dataset_all) - set(['ID', 'y']))
dataset__train_set = dataset_all[dataset_train_columns].iloc[:25317,:]
#dataset__train_y_set = dataset_all.iloc[:25317,['y']]
dataset__train_y_set = dataset_all['y'].iloc[:25317].T
dataset__test_set = dataset_all[dataset_train_columns].iloc[25317:,:]
```

```python
x_train, x_val, y_train, y_val = train_test_split(dataset__train_set, dataset__train_y_set, test_size=0.2, random_state=7)
```

```python
type(y_val.values)
```

```python
params = {
    'booster': 'gbtree',
    'eta': 0.05,
    'max_depth': 9,
    'objective': 'binary:logistic',
    'eval_metric':'auc',
    'seed':7
}
train_data = xgb.DMatrix(x_train, label=y_train)
val_data = xgb.DMatrix(x_val, label=y_val)
test_data = xgb.DMatrix(dataset__test_set)
#xgbc = xgb.XGBClassifier(**params)
#xgbc.fit(x_train, y_train, eval_set=(x_val, y_val), eval_metric='auc', early_stopping_rounds=50, verbose=1)
watchlist =[(train_data,'train'),(val_data,'val')]
```

```python
bst = xgb.train(params, train_data, 2000, watchlist, early_stopping_rounds=20, verbose_eval=1)
```

```python
val_pred = bst.predict(test_data)
```

```python
result_pd = pd.DataFrame(data=dataset_all.iloc[25317:,:].ID, columns=['ID'])
result_pd['pred'] = val_pred
result_pd.to_csv('result_xgb_20190326_standardscaler.csv', index=False)
```

```python
xgb_model = xgb.XGBClassifier(max_depth=11, learning_rate=0.05, n_estimators=300, verbosity=1, objective='binary:logistic', booster='gbtree', n_jobs=3, random_state=7)
xgb_model.fit(x_train, y_train, eval_set=[(x_val, y_val)], eval_metric='auc', early_stopping_rounds=50, verbose=1)
```
